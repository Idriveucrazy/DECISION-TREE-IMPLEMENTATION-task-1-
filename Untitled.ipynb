```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.tree as tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
df = pd.read_csv("drug200.csv")
df.head()
```

```python
X = df[['Age','Sex','BP','Cholesterol','Na_to_K']].values
X[0:5]
```

**NOW WE CONVERT CATEGORICAL VALUES TO NUMERICAL USING LABEL ENCODER**

```python
from sklearn import preprocessing
```

```python
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1])
```

```python
le_BP = preprocessing.LabelEncoder()
le_BP.fit(['LOW','NORMAL','HIGH'])
X[:,2] = le_BP.transform(X[:,2])
```

```python
le_chol = preprocessing.LabelEncoder()
le_chol.fit(['NORMAL','HIGH'])
X[:,3] = le_chol.transform(X[:,3])
```

```python
X[0:5]
```

```python
Y = df['Drug']
Y[0:5]
```

**NOW WE TRAIN AND TEST**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=3)
```

```python
drugTree = DecisionTreeClassifier(criterion = "entropy",max_depth = 4)
drugTree
```

```python
drugTree.fit(X_train,Y_train)
```

**PREDICTION**

```python
predTree = drugTree.predict(X_test)
print(predTree[0:5])
```

**EVALUATION**

```python
print("Decision Tree's Accuracy:",metrics.accuracy_score(Y_test,predTree))
```

**VISUALIZATION USING MATPLOT**

```python
tree.plot_tree(drugTree, feature_names=['Age','Sex','BP','Cholesterol','Na_to_K'], class_names=df['Drug'].unique(), filled=True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(Y_test, predTree)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=df['Drug'].unique())
disp.plot()
plt.show()
```
